```python
#loading the dataset
import pandas as pd
df = pd.read_csv('/train.csv')
print(df)
```

```python
#basic info
print(df.shape)
print(df.info())
print(df.describe())
#tells us how big the data is and how each colum looks like
```

```python
#checking missing values
print(df.isnull().sum())
#find where data is missing
```

```python
#value counts of important columns
print(df['Sex'].value_counts())
print(df['Pclass'].value_counts())
print(df['Embarked'].value_counts())
# .value_count helps in knowing counts of columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#countplot for categorical column
sns.countplot(df['Sex'])
plt.show()

sns.countplot(df['Pclass'])
plt.show()

sns.countplot(df['Embarked'])
plt.show()

#0bservation- if more males,more may have died.
```

```python
#distribution of age using histogram
sns.histplot(df['Age'].dropna(),kde=True)
plt.title('distribution of age')
plt.show()

#distribution of fare
sns.histplot(df['Fare'].dropna(),kde=True)
plt.title('distribution of fare')
plt.show()

#observations-1. most passengers are young adults
 #             2. most fares fall into a certain range, say between 0 and 50
```

```python
#survival rate by gender
sns.barplot(x='Sex',y='Survived',data=df)
plt.title('survival rate by gender')
plt.show()

#survival rate by passenger class
sns.barplot(x='Pclass',y='Survived',data=df)
plt.title('survival rate by passenger class')
plt.show()

#observation-1. more females survived than males
#            2. passenger class 1 survived more than other class
```

```python
print(df.columns)
```

```python
##boxplot for fare vs passenger class
sns.boxplot(x='Pclass',y='Fare',data=df)
plt.title('fare paid by different classes')
plt.show()

#observation - 1st class paid higher fares
```

```python
#heatmap
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.title('correlation matrix')
plt.show()

#observation-fare and pclass are negatively correlated
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.show()

#observations-helps find pattern between multiple columns
```
